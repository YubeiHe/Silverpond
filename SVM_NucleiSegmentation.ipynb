```python
from PIL import Image
from numpy import *
from svmutil import *
from svm import *

import numpy as np


scale = 0.3;  

def img2mat(I):
    pix = I.load()
    width = I.size[0]
    height = I.size[1]
    pixnum = I.size[0]* I.size[1]
    rgb_im = I.convert('RGB')  
    RGB=[]
    for x in range(width):
        for y in range(height):
            #rgb = pix[x, y]
            RGB.append(pix[x,y])
    feature = np.array(RGB)
    feature = feature.reshape((pixnum,3))
    feature = tuple(feature.tolist())
    
    return feature; 


def img2label(L):
    pix = L.load()
    pixnum = L.size[0]* L.size[1]
    L=L.convert('1')
    label=double(L)
    label =label.reshape(pixnum,1)
    label= array(label.tolist())

    return label;


#Train

#Read Image  
I=Image.open('x180_benign1_ccd.png');
#I=scipy.misc.imresize(I,scale)
feature = img2mat(I)
    
#Read label
L=Image.open(('x180_benign1_lab.png'));
#L=scipy.misc.imresize(L,scale)
label = img2label(L)

#Using SVM to train a model
prob=svm_problem(label,feature)
model=svm_train(prob);


#Test
#Read Image
II=Image.open('x180_benign2_ccd.png');

feature2 = img2mat(II)


#Read label
LL=Image.open(('x180_benign2_lab.png'));

label2 = img2label(LL)

#Test the model
svmlabel,svmacc,svmval= svm_predict(label2,feature2,model);
```
